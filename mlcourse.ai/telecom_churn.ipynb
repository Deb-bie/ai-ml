```python
import numpy as np
import pandas as pd

'''
pd.set_option: it is used for changing the various display settings in pandas. some examples are:
pd.set_option('display.max_rows', 10) - setting the maximum number of rows to display
pd.set_option('display.max_columns', 20) - setting the maximum number of columns to display
pd.set_option('display.max_colwidth', 100) - setting the maximum width of columns
pd.set_option("display.precision", 2) - setting the display precision for floating point numbers

'''
pd.set_option("display.precision", 2)
```

```python
DATA_URL = "https://raw.githubusercontent.com/Deb-bie/mlcourse.ai/main/data/"
df = pd.read_csv(DATA_URL + "telecom_churn.csv")


# df.head() is used to return the first few rows of a dataframe. By default, it returns the first 5 rows
# But you can specify a different number of rows as an argument. eg. df.head(25)
df.head() 
```

```python
# df.shape -  this is an attribute that returns a tuple that contains the dimensionality(rows and columns) of the dataframe.
# It returns (the number of rows, number of columns)
print(df.shape)
```

```python
# df.columns - it returns the columns in the dataframe
print(df.columns)
```

```python
# df.info() - it provides a concise summary of a dataframe.
# this summary includes the class type of the dataframe, index range and type, column names, 
# their data types, and the number of non-null values, memory usage of the dataframe
print(df.info())
```

```python
# astype - for converting a feature or a column's data type from one data type to another
df["Churn"] = df["Churn"].astype("int64")
df.head(5)
```

```python
# df.describe() - it shows the basic statistical characteristics of each numerical feature(int64 and float64 types):
# eg. number of non-missing values, mean, standard deviation, range, median, 0.25,0.75, 0.5 percentiles
df.describe()
```

```python
# In order to see statistics on non-numerical features, 
# one has to explicitly indicate data types of interest in the include parameter.
df.describe(include=["object", "bool"])
```

```python
# df.value_counts() - used to count the unique values in descending order for columns or features with categorical type 
# (type object and type bool)

df["Churn"].value_counts()
```

```python
'''
The churn rate is the percentage of subscribers who discontinue service subscriptions within a given time. 
2850 users out of 3333 are loyal; their Churn value is 0. That means they are still using the service
'''

# To calculate fractions, pass normalize=True to the value_counts function

df["Churn"].value_counts(normalize=True)
```

```python
'''
SORTING

A dataframe can be sorted by the value of one of the variables(columns).
Eg. we can sort by 'Total day charge' using acending=False. This wil then sort it in descending order.
if we don't use ascending=False, the default will be in ascending order
'''

df.sort_values(by="Total day charge", ascending=False).head()
```

```python
df.sort_values(by="Total day charge").head()
```

```python
# we can aslo sort by multiple columns
# In this example, "churn" is set to true and Total day charge is set to false
df.sort_values(by=["Churn", "Total day charge"], ascending=[True, False]).head()
```

```python
# we can also set ascending to just one value when dealing with multiple columns

df.sort_values(by=["Churn", "Total day charge"], ascending=False).head()
```

```python
'''Indexing and Retrieving Data'''

# To get a single column, you can use a DataFrame['Name'] construction. 
# Let’s use this to answer a question about that column alone: what is the proportion of churned users in our dataframe?

df["Churn"].mean()
```

```python
'''
BOOLEAN INDEXING

Boolean indexing with one column is also very convenient. The syntax is df[P(df['Name'])], 
where P is some logical condition that is checked for each element of the Name column. 
The result of such indexing is the DataFrame consisting only of the rows that satisfy the P condition on the Name column.
'''

# Question: What are the average values of numerical features for churned users?

# select_dtypes = for selecting a columns based on a specific data type

df.select_dtypes(include=np.number)[df["Churn"] == 1].mean()
```

```python
# How much time (on average) do churned users spend on the phone during daytime?

df["Total day minutes"][df["Churn"] == 1].mean()
# df[df["Churn"] == 1]["Total day minutes"].mean()
```

```python
# What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?

# df["Total intl calls"][df["Churn"] == 0][df["International plan"] == "No"].max()

df[(df["Churn"] == 0) & (df["International plan"] == "No")]["Total intl calls"].max()
```

```python
'''
DataFrames can be indexed by column name (label) or row name (index) or by the serial number of a row. 
The loc method is used for indexing by name, while iloc() is used for indexing by number.
'''
```

```python
# Example: “give us the values of the rows with index from 0 to 5 (inclusive) 
# and columns labeled from State to Area code (inclusive)”.

df.loc[0:5, "State":"Area code"]
```

```python
# In the second case, we say “give us the values of the first five rows in the first three columns” 
# (as in a typical Python slice: the maximal value is not included).

df.iloc[0:5, 0:3]
```

```python
# If we need the first or the last line of the data frame, we can use the df[:1] or df[-1:] construction:

# first row
df[:1]

# last row
df[-1:]
```

```python
'''Applying Functions to Cells, Columns and Rows'''
# To apply functions to each column, use apply():

df.apply(np.max)
```

```python
'''
The apply method can also be used to apply a function to each row. 
To do this, specify axis=1. Lambda functions are very convenient in such scenarios. 
'''

# For example, if we need to select all states starting with ‘W’, we can do it like this:
df[df["State"].apply(lambda state: state[0] == "W")].head()
```

```python
'''
The map method can be used to replace values in a column by passing a dictionary of the form 
{old_value: new_value} as its argument:
'''

d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)
df.head()
```

```python
'''
GROUPING

In general, grouping data in Pandas works as follows:
df.groupby(by=grouping_columns)[columns_to_show].function()

1. First, the groupby method divides the grouping_columns by their values. They become a new index in the resulting dataframe.

2. Then, columns of interest are selected (columns_to_show). 
   If columns_to_show is not included, all non groupby clauses will be included.

3. Finally, one or several functions are applied to the obtained groups per selected columns.
'''
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].mean()
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].describe()
```

```python
# Let’s do the same thing, but slightly differently by passing a list of functions to agg():
# agg() - used to perform multiple aggregation operations on a dataframe or series. 
# it allows you to apply one or more functions to one or more columns of a dataframe.


columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
'''
SUMMARY TABLES

Suppose we want to see how the observations in our dataset are distributed in the context of two variables – 
Churn and International plan. To do so, we can build a contingency table using the crosstab method:
'''
# crosstab() - used to compute a cross-tabulation of two or more factors. 
# It is used to analyze the relationship between two categorical values

pd.crosstab(df["Churn"], df["International plan"])
```

```python
# you can normalize the crosstab to show proportions instead of raw counts

pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

```python
# normalize by the index (rows)

pd.crosstab(df["Churn"], df["Voice mail plan"], normalize='index')
```

```python
# normalize by the columns

pd.crosstab(df["Churn"], df["Voice mail plan"], normalize='columns')
```

```python
# normalize by the entire table

pd.crosstab(df["Churn"], df["Voice mail plan"], normalize='all')
```

```python
# you can add margins (totals) to the rows and columns by settings the margins parameter to True:

pd.crosstab(df["Churn"], df["International plan"], margins=True)
```

```python
'''
We can see that most of the users are loyal and do not use additional services (International Plan/Voice mail).
This will resemble pivot tables to those familiar with Excel. And, of course, pivot tables are implemented in Pandas: 
the pivot_table method takes the following parameters:

values – a list of variables to calculate statistics for,
index – a list of variables to group data by,
aggfunc – what statistics we need to calculate for groups, e.g. sum, mean, maximum, minimum or something else.

Let’s take a look at the average number of day, evening, and night calls by area code:
'''

# pivot_table() - it allows you to summarize and aggregate data by one or more keys(columns) and perform various calculations
# such as sum, mean, count on the grouped data
# It is more versatile than crosstab() because it supports multiple aggregations and hierarchical indexing.


df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```

```python
'''
DataFrame Transformations

Like many other things in Pandas, adding columns to a DataFrame is doable in many ways.
'''
# For example, if we want to calculate the total number of calls for all users, 
# let’s create the total_calls Series and paste it into the DataFrame:


total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
# loc parameter is the number of columns after which to insert the Series object
# we set it to len(df.columns) to paste it at the very end of the dataframe
df.head()
```

```python
# It is possible to add a column more easily without creating an intermediate Series instance:

df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
```

```python
'''
To delete columns or rows, use the drop method, passing the required indexes and the axis parameter

When axis=1 -> when you want to delete columns
axis=0 or nothing -> when you want to delete rows

The inplace argument tells whether to change the original DataFrame. 
inplace=False, the drop method doesn’t change the existing DataFrame and returns a new one with dropped rows or columns. 
inplace=True, it alters the DataFrame.

'''

# get rid of just created columns
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
# and here’s how you can delete rows
df.drop([1, 2]).head()
```

```python
# Let’s see how churn rate is related to the International plan feature. 
# We’ll do this using a crosstab contingency table and also through visual analysis with Seaborn

pd.crosstab(df["Churn"], df["International plan"], margins=True)
```
