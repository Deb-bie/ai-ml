```python
DATA_URL="https://raw.githubusercontent.com/Deb-bie/ai-ml/main/kaggle/finance/Finance_data.csv"
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%config InlineBackend.figure_format = 'retina'
```

```python
# plotting a graph around the gender
sns.countplot(x="gender", data=df)
```

```python
# gender and age
pd.crosstab(df["gender"], df["age"], margins=True)
```

```python
# how many males or females have investment avenues
pd.crosstab(df["gender"], df["Investment_Avenues"], margins=True)
```

```python
sns.countplot(x="gender", hue="Investment_Avenues", data=df)
```
